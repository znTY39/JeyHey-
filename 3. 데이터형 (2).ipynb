# 1. 숫자
### 컴퓨터가 가장 잘 이해하는 표현은 숫자이며, 수의 체계는 다음과 같음

### - 파이썬에서는 수 체계의 모든 표현을 사용하지 않고 실수(float)과 정수(integer) 그리고 허수(complex number)만 사용함

#### 
####  아래와 같이 숫자 3을 출력하였음
#### 숫자 3은 수 체계에서 자연수에 속함

```python
print(3)
```

### 
### - 파이썬에서 데이터형을 보기 위한 함수인 type() 함수가 있음
#### type() 괄호 안에 데이터를 넣고 print() 함수로 출력하면 파이썬에서 어떤 데이터형으로 인식하는지 알 수 있음
#### 파이썬에서는 숫자 3을 자연수보다 위에 있는 정수로 분류함

```python
print(type(3))
```

#### <class 'int'> : class 는 어떤 분류에 속하는지를 나타내고, 'int' 는 integer 정수의 줄임말을 표현함
#### 즉 3 은 파이썬에서 정수 int 데이터형에 속함

#### 
#### 그렇다면 정수가 아닌 다른 수 3.6 을 type() 에 넣어 데이터형을 확인해보자

```python
print(type(3.6))
```

#### <class 'float'> : 'float' 은 실수를 표현함
#### 즉 3.6 은 실수 float 데이터형에 속함

#### 
#### 문제 : 3/5 과 0.5 과 -5 을 type() 을 사용하여 데이터형을 출력해보세요

#### 
#### 0 과 0.0 의 데이터형을 출력해보면 데이터형이 각각 int와 float으로 다르다.
#### * 파이썬에서는 소수점이 붙으면 float 데이터형으로 분류한다.

```python
print(type(0))
print(type(0.0))
```

#### 
### - 파이썬의 int, float 데이터형은 숫자이므로 덧셈,뺄셈,곱셈,나눗셈 등의 연산을 할 수 있음
#### 하지만 어떤 연산을 수행하는가에 따라 데이터형의 결과가 다르게 나올 수 있음
#### int, float 의 사칙연산 결과 데이터형 표 :


#### int 데이터형끼리 사칙연산 결과
#### * 나눗셈의 경우 소수점이 나올 수 있어 float 으로 데이터형을 바꿈

```python
print(type(3 + 2))
print(type(3 - 2))
print(type(3 * 2))
print(type(3 / 2))
```

#### int와 float 데이터형의 사칙연산 결과
#### * int와 float 의 사칙연산 결과는 모두 float 으로 인식

```python
print(type(3 + 2.0))
print(type(3 - 2.0))
print(type(3 * 2.0))
print(type(3 / 2.0))
```

#### float 데이터형끼리 사칙연산 결과

```python
print(type(3.0 + 2.0))
print(type(3.0 - 2.0))
print(type(3.0 * 2.0))
print(type(3.0 / 2.0))
```

#### 즉, 연산결과는 float이 우세함

# 2. 문자열
### - 파이썬이 이해할 수 있는 데이터형 중 하나가 문자열(String) 임
### - 문자열은 글자들의 집합을 의미함
### - 파이썬은 모든 표현에 쌍따옴표("") 혹은 작은따옴표('') 을 감싸면 문자열로 인식함

#### 
#### 다양한 문자열 예시
#### " " 안에 들어오면 무슨 표현이든 문자열 데이터형으로 간주함

```python
print(type("고영민"))
print(type("python"))
print(type("3"))
print(type("2.5"))
print(type("%*@(*@)"))
print(type("하늘을 날 것 같아요!"))
```

#### 
#### 문제 : 여러분의 이름을 출력하고, type() 을 사용하여 데이터형을 출력해보세요

#### 
#### <class 'str'> 에서 str 은 문자열 string의 줄임말임
#### * 파이썬에서 int 인 3과 float 인 2.5 도 쌍따옴표"" 안에 들어가면 모두 문자열로 인식함
#### * 특수기호를 섞은 것도 쌍따옴표 안에 들어가면 모두 문자열 str로 인식함
#### * 띄어쓰기도 쌍따옴표 안에 들어가면 모두 문자열로 str로 간주함
#### * 파이썬에서는 작은따옴표와 큰따옴표의 구분이 없음

```python
print(type('2'))
print(type("2"))
```

# 3. 불(Bool) 또는 불린(Boolean) 데이터형 : 참(True) 과 거짓(False)

### - 불(Bool) 데이터형은 참(True) 이나 거짓(False)만 저장할 수 있는 데이터 형식임
### - 논리형 이라고 부르기도함
#### * 불의 어원은 수학자 겸 논리학자인 조지 불(George Boole)의 이름에서 따온 것임

```python
print(type(True))
```

```python
True
```

```python
print(type(False))
```

#### * 불 데이터형은 조건문인 if문과 주로 같이 사용됨
#### * 나중에 배울 비교, 논리연산자와 밀접한 관련이 있음

```python
print(100 > 10)
```

```python
print(100 < 10)
```

#### * 또한 컴퓨터는 True 를 숫자 1 로, False 를 숫자 0으로도 인식함
#### 즉, 사칙연산이 가능하다는 의미임

```python
print(True+True)
```

```python
print(True + False)
```

```python
print(False + False)
```
